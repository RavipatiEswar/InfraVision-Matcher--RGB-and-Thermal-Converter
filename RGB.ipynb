```python
import os
from PIL import Image
import io

# Creating the output folder that to automatically store the images 
os.makedirs("Output_folder", exist_ok=True)

pairs = []
for t_file in upload_thermal.value:
    # to get the base name of the image without extension
    filename = t_file['name']  
    base = filename.rsplit('_', 1)[0]  
    
    # Finding the matching RGB file with same prefix
    match = next((r for r in upload_rgb.value if r['name'].startswith(base)), None)
    if match:
        t_img = Image.open(io.BytesIO(t_file['content']))
        r_img = Image.open(io.BytesIO(match['content']))
        pairs.append((filename, t_img, r_img))
        print(f"Paired {t_file['name']} with {match['name']}")

# Aligning  and saving  with correct naming
for filename, t_img, r_img in pairs:
    # Resizing the  thermal image  to match RGB image  size exactly 
    t_aligned = t_img.resize(r_img.size, Image.BILINEAR)
    
    # Replac ing the "_T" with "_AT" before extension
    new_name = filename.replace("_T", "_AT")
    save_path = os.path.join("Output_folder", new_name)
    
    # Saving the  aligned thermal image at the dir'Output_folder'
    t_aligned.save(save_path)
    print(f"Saved aligned thermal image: {save_path}")
```
